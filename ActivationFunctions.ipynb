```python
import numpy as np
import matplotlib.pylab as plt
```

## ステップ関数

$
h(x)=\begin{cases}
  0\quad(x \leqq 0) \\
  1\quad(x > 0)
\end{cases}
$

```python
def step_function(x):
    return np.array(x > 0, dtype = np.int)

x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)

plt.plot(x, y)
plt.show()
```

## シグモイド関数
$
h(x) = \frac{1}{1 + e^{-x}}
$

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.plot(x, y)
plt.show()
```

## ReLU関数

$
h(x)=\begin{cases}
  0\quad(x \leqq 0) \\
  x\quad(x > 0)
\end{cases}
$

```python
def relu(x):
    return np.maximum(0, x)

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)

plt.plot(x, y)
plt.show()
```

## ソフトマックス関数

$
y_k = \frac{e^{a_k}}{\sum_{i=1}^ne^{a_i}}
$

```python
def softmax(a):
  exp_a = np.exp(a)
  sum_exp_a = np.sum(exp_a)
  y = exp_a / sum_exp_a
  return y

x = np.arange(-5, 5, 0.1)
y = softmax(x)
plt.plot(x, y)
plt.show()
```

## 恒等関数

$
y=x
$

```python
def identity_function(x):
    return x
x = np.arange(-5, 5, 0.1)
y = identity_function(x)
plt.plot(x, y)
plt.show()
```

## tanh関数
$
y = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$

```python
def tanh(x):
    numer = np.exp(x) - np.exp(x * -1)
    denom = np.exp(x) + np.exp(x * -1)
    return numer / denom
x = np.arange(-5, 5, 0.1)
y = tanh(x)
plt.plot(x, y)
plt.show()
```
